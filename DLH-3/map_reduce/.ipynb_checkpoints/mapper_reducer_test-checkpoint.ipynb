```python
import numpy as np
```

# mapper

```python
with open('keywords_after_mapper.csv', 'w') as new_file:
    keywords=[]
    with open('keywords.csv') as f:
        for i, line in enumerate(f):
            keyword = line.strip().split(',')[:-1]
            keywords.append(keyword)

    list_of_keywords = np.array(keywords).flatten()

    words = []
    for i in list_of_keywords:
            word = i.strip().split(' ')
            words.append(word)

    list_of_words = np.array(words).flatten()

    flat_list_of_words = [word for sublist in list_of_words for word in sublist]

    for i in flat_list_of_words:
        new_file.write(f"{i},1\n")
```

# Shuffle

cat keywords_after_mapper.csv | sort > keywords_after_mapper_sorted.csv

# Reducer

```python
prev_word = None
counter = 0

with open('keywords_after_mapper_sorted.csv') as f:
    for i, line in enumerate(f):
        word, one = line.strip().split(',') 
        one = int(one)
        
        if prev_word:
            if prev_word == word:
                counter += one
            else:
                print(prev_word, counter)
                prev_word = word
                counter = one
        else:
            prev_word = word
            counter = one

print(prev_word, counter)
```
